```python
def extract_episode(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            break
        if i == "(":
            check = True
            continue
            
        if check ==True:
            
            data += i
            
    return data
```

```python
extract_episode("Fullmetal Alchemist: BrotherhoodTV (64 eps)")
```

```python
# Q1 : Make a new column for episode count

df["Episode"]= df["Title"].apply(extract_episode)
```

```python
df["Episode"]=df["Episode"].str.replace(" eps","")
```

```python
df["Episode"]=df["Episode"].astype(int)
```

```python
# make a column of time stamp
df.loc[0]["Title"]
```

```python
def extraction_time(txt):
    for i in range(len(txt)):
        if txt[i] == ")":
            data= txt[i+1:i+20:1]
    return data
```

```python
df["Time Stamp"] = df["Title"].apply(extraction_time)
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

df['Months'] = df['Time Stamp'].apply(calculate_total_months)
```
